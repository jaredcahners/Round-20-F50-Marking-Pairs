## What Kicker-Marker Pairs Have Tallied the Most Inside 50 Marks through AFL Round 20?

```python
#import the libraries

import pandas as pd
import numpy as np
```

```python
#read the csv
r20 = pd.read_csv('AFL Play by play to r20.csv')
```

```python
pd.set_option('display.max_rows', None)

r20.head(300)
```

```python
#add a 'Possession' column to determine who has possession and a 'Poss_id' column to give each possession an id

#turning the dataframe into a dictionary makes this happen much faster

r20_dict = r20.to_dict('records')

poss = None
quarter = 0
Poss_list = []
Poss_id_list = []
n=0
for row in r20_dict:
    if row['Quarter'] != quarter:
        poss = None
        Poss_id_list.append(None)
        quarter = row['Quarter']
        row['Possession'] = poss
        Poss_list.append(poss)
    else:
    
        if row['Stat'] in ['Poss Gain', 'Centre Clearance', 'Stoppage Clearance', 'Kick In', 'Contested Possession']:
            poss = row['Team_status']
            row['Possession'] = poss
            Poss_list.append(poss)

        elif row['Stat'] in ['Hitout', 'Hitout To Advantage']:
            poss = None
            row['Possession'] = poss
            Poss_list.append(poss)
            
        elif row['Stat'] == 'Turnover':
            if poss == 'home':
                poss = 'away'
            else:
                poss = 'home'
            row['Possession'] = poss
            Poss_list.append(poss)
            
        elif row['Stat'] not in ['Hitout', 'Hitout To Advantage', 'Poss Gain', 'Centre Clearance', 'Stoppage Clearance', 'Kick In', 'Contested Possession', 'Goal', 'Behind', 'Turnover']:
            row['Possession'] = poss
            Poss_list.append(poss)
        
        elif row['Stat'] in ['Goal', 'Behind']:
            row['Possession'] = poss
            Poss_list.append(poss)
            poss = None

#possession counter        
        
        if row['Possession'] == None:
            Poss_id_list.append(None)

        elif row['Possession'] == Poss_list[-2] or row['Possession'] == 'score':
            Poss_id_list.append(Poss_id_list[-1])
        else:
            n+=1
            Poss_id_list.append(n)


r20 = pd.DataFrame(r20_dict)          
            

r20['Poss_id'] = Poss_id_list

```

```python
r20.head(50)
```

```python
#create a name column from first name and last name

r20['Player'] = r20['First_name'] + ' ' + r20['Last_name']
```

```python
#narrow down the data to kicks and forward 50 marks

Kicks_f50marks = r20[(r20['Stat'] == 'Kick') | (r20['Stat'] == 'F 50 Mark') ].copy().reset_index()
Kicks_f50marks.head(30)
```

```python
#Let's take the forward markers, the last kicker, their team and the possession id and create a new dataframs: 'F50'.

kickers = []
markers = []
teams = []
poss_id = []
for ind, row in Kicks_f50marks.iterrows():
    if row['Stat'] == 'F 50 Mark':
        kickers.append(Kicks_f50marks.loc[ind-1, 'Player'])
        markers.append(Kicks_f50marks.loc[ind, 'Player'])
        teams.append(Kicks_f50marks.loc[ind, 'Team'])
        poss_id.append(Kicks_f50marks.loc[ind, 'Poss_id'])

F50 = pd.DataFrame(list(zip(kickers, markers, teams, poss_id)), columns = ['Kicker', 'Marker', 'Team', 'Poss_id'])
```

```python
F50.head()
```

```python
#Now we go back to the old dataframe and put in a points column based on the stats that are scores.
#Then will group by the possession to get the points scored each possession
r20['Points'] = np.where(r20['Stat'] == 'Goal', 6, np.where(r20['Stat'] == 'Behind', 1, 0))
r20_points = r20.groupby('Poss_id', as_index = False)['Points'].sum().copy()
```

```python
r20_points.head(20)
```

```python
#Next let's make a dictionary that tells us how many points come from each possession
poss_score_dict = {}
for ind, row in r20_points.iterrows():
    poss_score_dict[row['Poss_id']] = row['Points']
```

```python
# Going through the F50 dataframe let's make a list of the points scored in those possessions from the dictionary.
Points_list = [poss_score_dict[i] for i in F50['Poss_id']]
```

```python
#And we'll make that list into a column
F50['Points'] = Points_list
```

```python
F50.head() 
```

```python
#Now let's get a count of the total F50 marks for each pair and the average amount of points scored during those possessions.
F50 = F50.groupby(['Kicker', 'Marker', 'Team'], as_index = False).agg({'Poss_id' : 'count', 'Points' : 'mean'})
```

```python
F50.rename(columns = {'Poss_id' : 'Count'}, inplace = True)
```

```python
F50.head()
```

```python
#Now let's sort by the number of forward 50 marks
F50.sort_values('Count', ascending = False).head(90)
```

```python
#Some games were left out of this data set.
#I want to have a row that estimates the F50 counts if those games were included.
#I did that by adding each pair's per game average, or, in Carlton's case, twice that.

# Round 4 - North Melbourne vs Adelaide
# Round 4 - Port Adelaide vs Richmond
# Round 7 - Sydney vs Geelong
# Round 8 - Brisbane vs Fremantle
# Round 8 - Western Bulldogs vs Carlton
# Round 9 - Melbourne vs Carlton

# North Melbourne
# Adelaide Crows
# Port Adelaide
# Richmond
# Sydney Swans
# Geelong Cats
# Brisbane Lions
# Fremantle
# Western Bulldogs
# Carlton *2
# Melbourne


for ind, row in F50.iterrows():
    if row['Team'] in ['North Melbourne', 'Adelaide Crows', 'Port Adelaide',' Richmond', 'Sydney Swans', 'Geelong Cats', 'Brisbane Lions', 'Fremantle', 'Western Bulldogs', 'Melbourne']:
        F50.loc[ind, 'Adjusted Count'] = F50.loc[ind, 'Count']*19/18
    elif row['Team'] == 'Carlton':
        F50.loc[ind, 'Adjusted Count'] = F50.loc[ind, 'Count']*19/17
    else:
        F50.loc[ind, 'Adjusted Count'] = F50.loc[ind, 'Count']
       
```

```python
#Let's round the numbers and sort by the new column

F50['Adjusted Count'] = F50['Adjusted Count'].round(2)
F50['Points'] = F50['Points'].round(2)
F50.sort_values('Adjusted Count', ascending = False).head(20)
```

```python
#Now, for presentation I want to reorder the columns, and reindex the dataframe starting with 1 instead of 0

F50 = F50[['Kicker', 'Marker', 'Team', 'Count', 'Adjusted Count', 'Points']]
F50 = F50.sort_values('Adjusted Count', ascending = False).head(20)
F50 = F50.reset_index()
F50.index = F50.index + 1
F50 = F50.drop(columns = 'index')
F50.head(20)
```
